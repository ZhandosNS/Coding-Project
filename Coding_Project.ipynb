```python
import pandas as pd

file_path = r'C:\Users\mrrai\OneDrive\Рабочий стол\python\morg2014.csv'

df = pd.read_csv(file_path)

print(df.head())


#checking the types of variables
pd.set_option('display.max_rows', None)  
pd.set_option('display.max_columns', None)  

print(df.dtypes)
```

```python
#8 creating summary statistics

columns_to_analyze = ['earnwke', 'uhourse', 'age']
summary_stats = df[columns_to_analyze].describe()

print(summary_stats)
```

```python
#9 creating a histogram of weekly earnings

import matplotlib.pyplot as plt

df['earnwke'].hist(bins=20, edgecolor='black')

plt.title('Histogram of weekly earnings')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.show()
```

```python
#5 checking the number of missing values in selected variables

columns_to_check = ['earnwke', 'age', 'uhourse']

missing_values = df[columns_to_check].isnull().sum()

print(missing_values)

# substituting missing values with the mean in "earnwke"

df['earnwke'] = df['earnwke'].fillna(df['earnwke'].mean())

missing_values_column = df['earnwke'].isnull().sum()

print(f"Missing values in 'earnwke': {missing_values_column}")
```

```python
#6 keeping observations only with "earnwke" higher than 500

filtered_df = df[df['earnwke'] >= 500]
print(filtered_df.head())


import pandas as pd
import numpy as np
# taking a logarithm of weekly earnings
filtered_df['log_earnings'] = np.log(filtered_df['earnwke'])
print(filtered_df.head())

# dropping "grade92" variable
filtered_df1 = df.drop(columns=['grade92'])
print(filtered_df1.head())
```

```python
#2 creating a new variable "earnings - age"

for index, row in filtered_df1.iterrows():
    filtered_df1.at[index, 'earnings - age'] = row['earnwke'] - row['age']

print(filtered_df1.head())
```
